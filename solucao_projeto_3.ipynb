```python
import copy
```

```python
class Mapa:

    def __init__(self, mapa):
        # Importante usar o deepcopy para não alterar as variáveis passadas
        self.mapa_original = copy.deepcopy(mapa) # Variável usada no método representativo
        self.mapa = copy.deepcopy(mapa) # Variável usada na varredura que será modificada


    def __conta_tamanho_rio__(self, x, y, tamanho = 0):
        # Criei uma variável auxiliar pra não ter que escrever self toda hora
        mapa = self.mapa

        tamanho += 1
            
        # Visita o ponto acima
        if y > 0 and mapa[y - 1][x] == 1:
            mapa[y - 1][x] = 'Visitado'
            tamanho = self.__conta_tamanho_rio__(x, y - 1, tamanho)


        # Visita o ponto abaixo
        if y + 1 < len(mapa) and mapa[y + 1][x] == 1:
            mapa[y + 1][x] = 'Visitado'
            tamanho = self.__conta_tamanho_rio__(x, y + 1, tamanho)


        # Visita o ponto à direita
        if x + 1 < len(mapa[0]) and mapa[y][x + 1] == 1:
            mapa[y][x+1] = 'Visitado'
            tamanho = self.__conta_tamanho_rio__(x + 1, y, tamanho)


        # Visita o ponto à esquerda
        if x > 0 and mapa[y][x - 1] == 1:
            mapa[y][x-1] = 'Visitado'
            tamanho = self.__conta_tamanho_rio__(x - 1, y, tamanho)

        return tamanho
        

    def acha_rio(self):

        tamanho_rios = []

        for y in range(len(self.mapa)):
            for x in range(len(self.mapa[0])):
                if self.mapa[y][x] == 1:
                    self.mapa[y][x] = 'Visitado'

                    # Começa a varredura a partir do primeiro ponto = 1 encontrado
                    tamanho_rios.append(self.__conta_tamanho_rio__(x, y))

        self.tamanho_rios = sorted(tamanho_rios)

    def __repr__(self):

        texto = ''
        texto += 'Contagem dos rios do mapa:\n\n'

        for y in range(len(self.mapa_original)):
            texto += '| '
            for x in range(len(self.mapa_original[0])):
                texto += str(self.mapa_original[y][x]) + ' '
            texto += '|\n'

        if self.tamanho_rios == []:
            texto += '\nNão encontrei nenhum rio!'
        else:
            texto += f'\nEncontrei {len(self.tamanho_rios)} rios: {self.tamanho_rios}\n'

        return texto


```
